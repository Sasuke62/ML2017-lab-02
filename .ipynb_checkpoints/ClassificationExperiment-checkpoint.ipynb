```python
#NAG（Nesterov accelerated gradient）
#RMSProp
#AdaDelta
#Adam（Adaptive Moment Estimation）
import numpy as np
import random
import matplotlib.pyplot as plt
 
from sklearn.externals.joblib import Memory
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_svmlight_file
from numpy import genfromtxt
```

```python
def get_data():
    data = load_svmlight_file('../ML2017-lab-1/australian_scale.txt')
    return data[0], data[1]
```

```python
def loss(X, y, W,lamda):
    num_records, num_features = np.shape(X)
    hinge_loss_train = 1.0 / float(num_records) * \
                                np.sum(np.max([np.zeros((num_records,1)),1 - y \
                               * (X.dot(W))],axis=0)) + 0.5 * lamda \
                                * W.transpose().dot(W)
    return hinge_loss_train[0][0]
```

```python
def gradient_descent(X, y, W,lamda):
    num_records, num_features = np.shape(X)
    hinge_loss_gradient = np.max([np.zeros((num_records,1)),1 - y * (X.dot(W))],axis = 0)
    indicator = np.zeros((num_records, 1))
    indicator[np.nonzero(hinge_loss_gradient)] = 1
    gradient_W = - 1.0 / float(num_records) * X.transpose().dot(y \
                * indicator).sum(axis = 1).reshape((num_features, 1)) + lamda * W
    return gradient_W, V
```

```python
def update_gradient_descent(X, y, W, lamda):
    gradient_W = gradient_descent(X, y, W,lamda)
    gama = 0.9
    W = gama * W - learning_rate * gradient_W
    return W
```

```python
def update_NAG(X, y, W, V, lamda):
    gama = 0.9
    gradient_W = gradient_descent(X, y, W - gama * V,lamda)
    V = gama * V + lamda * gradient_W
    W = W - V
    return W ,V
```

```python
X,y = get_data()
y = y.reshape((len(y), 1))
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size = 0.3, random_state = 1)
```
